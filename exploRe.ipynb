<b>Problem 1: Define two variables 'sales1' and 'sales2' and assign 12 values to them. sales2 is assignes values at random using Poission distribution having mean 34.</b>

```r
#Assign values to sales1 and sales2
sales1 <- c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2 <- rpois(12,34)     # random numbers, Poisson distribution, mean at 34, 12 numbers

#Plot sales1 and sales2 by Month 
par(bg="cornsilk")
plot(sales1, col="blue", type="o", ylim=c(min(c(sales1,sales2))-10,max(c(sales1,sales2))+10), xlab="Month", ylab="Sales" )
title(main="Sales by Month")
lines(sales2, type="b", pch=22, lty=2, col="red")
grid(nx=NULL, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

```r
# load the file 
if(!file.exists('./data_files/summary_data.csv')) {
  theurl <- getURL("https://www.cdc.gov/flu/weekly/weeklyarchives2017-2018/data/whoAllregt_cl04.html", .opts = list(ssl.verifypeer = FALSE) )
  raw_table <- readHTMLTable(theurl)
  clean_table <- list.clean(raw_table, fun = is.null, recursive = FALSE)
  summary_data <-clean_table[[1]]
  row.names(summary_data) <- NULL
  write.csv(summary_data,file="./data_files/summary_data.csv",row.names = FALSE)
  
}
# read in  the data in file
summary_data <- read.csv('./data_files/summary_data.csv')
#print(summary_data)

# preprocessing and removing spaces in column names
summary_data[,1] <- as.character(summary_data[,1])
names(summary_data) <- gsub("\\.", "", names(summary_data))

# plot the graph
ggplot(summary_data ) + 
  scale_y_continuous(sec.axis = sec_axis(~./max(summary_data$TotalA)*max(summary_data$XPositive), name ='Percent Positives')) +

geom_bar(stat='identity',aes( x= Week,y = TotalA, fill = 'Total A'), color = 'black') +
  geom_bar(stat='identity',aes( x= Week,y = TotalB, fill = 'Total B'), color = 'black') +
  labs(x = 'Week', y = 'Number of positive specimens', 
       caption = 'Based on data from https://www.cdc.gov/flu/weekly/weeklyarchives2017-2018/Week04.htm',
       title = 'Influenza Positive Tests Reported to CDC by U.S. Clinical laboratories, National Summary, 2017-2018 Season') +

geom_line(stat = 'identity', aes(color = '1', x = Week, y =XPositive*max(summary_data$TotalA)/max(summary_data$XPositive), group = 3 ), size = 1) +
geom_line(stat = 'identity', aes(color ='2',x = Week, y =PercentPositiveA*max(summary_data$TotalA)/max(summary_data$XPositive), group = 1), linetype = 'dashed', size = 1) +
geom_line(stat = 'identity', aes(color ='3',x = Week, y =PercentPositiveB*max(summary_data$TotalA)/max(summary_data$XPositive), group = 2 ), linetype = 'dashed', size = 1) +
  
scale_color_discrete(name = 'Legends', labels = c("Percent Positive", "% positive flu A", "% positive flu B") ) +
theme(axis.text.x = element_text(face="bold", color="black", size=9, angle=60),
      axis.text.y = element_text(face="bold", color="black", size=9),
      legend.position = c(0.2,0.7),
      panel.background = element_rect(fill = 'white', colour = 'black'))
```
